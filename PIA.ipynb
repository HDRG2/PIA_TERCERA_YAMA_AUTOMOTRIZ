```python
import pandas as pd
```

```python
dTiposDeDatos ={
    'Vin' : object,
    'NumeroInventario': int,
    'CostoCompra': float,
    'CostoActual': float,
    'Año': int,
    'Marca': object,
    'Modelo': object,
    'Version': object,
    'Color': object,
    'Interior': object,
    'Estado': object,
    'TipoCompra': object,
    'FechaCompra': object,
    'Placas': object,
    'NuevoUsado': object,
    'Ubicacion': object,
    'Nada': object
}
dTiposDeDatos
```

```python
dfInvApodaca = pd.read_csv('INVAPODACA.txt', sep = '|', dtype = dTiposDeDatos)
dfInvApodaca
```

```python
dfInvEscobedo = pd.read_csv('INVESCOBEDO.txt', sep = '|', dtype = dTiposDeDatos)
dfInvEscobedo
```

```python
dfInvMonterrey = pd.read_csv('INVMONTERREY.txt', sep = '|', dtype = dTiposDeDatos)
dfInvMonterrey
```

```python
dfInvSanNicolas = pd.read_csv('INVSANNICOLAS.txt', sep = '|', dtype = dTiposDeDatos)
dfInvSanNicolas
```

```python
dfInv = pd.concat([dfInvApodaca, dfInvEscobedo, dfInvMonterrey, dfInvSanNicolas], axis=0)
dfInv
```
